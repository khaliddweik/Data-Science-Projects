

### -- Dataset: https://www.kaggle.com/rio2016/olympic-games?select=athletes.csv 

### “This dataset consists of the official statistics on the 11,538 athletes at the 2016 Olympic Games in Rio de Janeiro. The athletes file includes the name, nationality (as a three letter IOC country code), gender, age (as date of birth), height in meters, weight in kilograms, sport, and quantity of gold, silver, and/or bronze medals won for every Olympic athlete at Rio.”

### -- Strategy used to explore your data and draw visualizations (what variables you used to plot visuals, and what other variables you ignored)

#### We ignored id and name because we felt that they were an unnecessary string variables for us to visualize in our graphs. They do not provide an overall picture to the story we are trying to tell
#### We ignored sex because we wanted to compare all atheltes, regardless of their gender.
#### Lastly we ignored date of birth, we believe that the age does not affect the performance of the athletes in their respective sports, and the medals they received. 
#### We used nationality, date of birth, height, weight, gold, silver, and bronze medals, and sports to visualize the data set. We believe that these variables have the most story to tell throughout the visualization.


### -- How you dealt with missing values in the data (explain why you think your method is correct)
####  Even though our missing values were less than usual, we used the mean to replace null values in the height and weight columns, while we used the mode to replace the date of birth column. The height and weight columns were floats and the best way to replace the missing values was using the mean of them, while the date of birth column is an object and the mode was the best way to replace the missing value.
#### Expanded on extra visualization used. We have an extra visualization (a bubble chart), as our missing values were not enough to completely satisfy the project requirement.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
```

```python
athletes = pd.read_csv("athletes.csv")
pd.read_csv("athletes.csv")
```

```python
athletes.info()
```

```python
athletes.isnull().sum()
```

```python
for column in ['height']:
    athletes[column].fillna(athletes[column].mean(), inplace = True)
    
for column in ['weight']:
    athletes[column].fillna(athletes[column].mean(), inplace = True)
    
for column in ['dob']:
    athletes[column].fillna(athletes[column].mode()[0], inplace = True)
    
athletes.isna().sum()
```

### -- Pie Chart

#####  The pie chart represents the number, in this case the percentages, of each sport that was played in the Olympics out of the total. Charts like these usually show how each slice of the pie is relative to the size of the group as a whole. As seen, athletics, fencing, and taekwondo were the top 3 sports coming in at 20.5%, 12.5%, and 5.3% respectively. The sports with the least amount of players was sailing, tennis, and gymnastics which came at 1.0%, 1.0%, and 0.6% respectively.

```python
##Pie Chart


data = pd.read_csv('athletes.csv')    #reading csv file and storing its content into data variable
sport = data['sport'].value_counts()   #counting the distinct number of sports
sport_name = data['sport'].unique()    #getting only unique sport name
plt.figure(figsize =(9,9))         # resizing the pie chart
plt.pie(sport, labels = sport_name,autopct='%1.1f%%',wedgeprops={'edgecolor': 'gray'})   #passing the value extracted from csv to pie chart
plt.title("Frequency of Sports")
plt.tight_layout()
plt.show()     #displaying the pie chart
```

### -- Boxplot

#####  This boxplot plots the height range for the top ten nations with the most athletes that competed in the 2016 Rio Olympics. In this visualization, you can see the quartiles for each of the top ten nations, with the outliers from most of them. For example, the US has the tallest person to compete, while Brazil has the shortest person to compete (in the top ten most common nationalities). On the other had, Japan does not have any outlier in their athletes’ heights that competed.

```python
#Boxplot

counts = collections.Counter(athletes.nationality)
top10countries = [elem for elem, _ in counts.most_common(10)]

df_top10 = athletes[athletes['nationality'].isin(top10countries)]

bp = sns.boxplot(y=df_top10['height'], x=df_top10['nationality'], data=athletes,
                 palette="colorblind", width = 0.75).set_title('Boxplot of the top 10 nationalities vs the height')

plt.grid(True)
plt.savefig('bp.png')
plt.show()
```

### -- Scatter Plot

#####  This scatter plot looks at the correlation between height and weight and if that affects the amount of gold medals won by an individual. In the plot you can see that there is a positive correlation between height and weight, the taller an individual is the heavier they are. There is also a color bar, the points that are blue shows that individuals won less gold medals whereas if they are a red color it shows that they won many gold medals. 

```python
#Scatter Plot

us = athletes[athletes.nationality == 'USA']
rus = athletes[athletes.nationality == 'RUS']

plt.figure(figsize=(16, 4.5))
plt.style.use('seaborn')
plt.scatter(us.height, us.weight, s= 25, c= us.gold, cmap='coolwarm', marker= 'o', edgecolor= 'black', linewidth= 1, alpha= 0.75)

cbar = plt.colorbar(ticks=[0, 1, 2, 3, 4, 5])
cbar.set_label('Gold Medals')
cbar.ax.set_yticklabels(['0', '1', '2', '3', '4', '5'])

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Relationship between Height, Weight and the Amount of Gold Medals - USA')

plt.grid(True)
plt.show()

plt.figure(figsize=(16, 4.5))
plt.scatter(rus.height, rus.weight, s= 25, c= rus.gold, cmap='coolwarm', marker= 'o', edgecolor= 'black', linewidth= 1, alpha= 0.75)

cbar = plt.colorbar(ticks=[0, 1, 2, 3, 4, 5])
cbar.set_label('Gold Medals')
cbar.ax.set_yticklabels(['0', '1', '>1'])

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Relationship between Height, Weight and the Amount of Gold Medals - Russia')

plt.grid(True)
plt.show()
```

### -- Line Plot

#####  The line plot usually displays changes over time but because our dataset did not have a variable like that we decided to illustrate the amount of medals won by each of the top 5 winningest nations. The gold line signified the gold medals, silver line for silver medals and bronze for the bronze medals. From the chart it looks like silver medals were the lowest for all of them while gold medals were the highest. 

```python
#Line Plot

us = athletes[athletes.nationality == 'USA']
rus = athletes[athletes.nationality == 'RUS']
gbr = athletes[athletes.nationality == 'GBR']
chn = athletes[athletes.nationality == 'CHN']
ger = athletes[athletes.nationality == 'GER']

usgold = us.gold.sum()
ussilver = us.silver.sum()
usbronze = us.bronze.sum()

rusgold = rus.gold.sum()
russilver = rus.silver.sum()
rusbronze = rus.bronze.sum()

gbrgold = gbr.gold.sum()
gbrsilver = gbr.silver.sum()
gbrbronze = gbr.silver.sum()

chngold = chn.gold.sum()
chnsilver = chn.silver.sum()
chnbronze = chn.bronze.sum()

gergold = ger.gold.sum()
gersilver = ger.silver.sum()
gerbronze = ger.bronze.sum()


plt.figure(figsize=(10, 4.5))
plt.style.use('seaborn')

names = ['USA', 'RUS', 'GBR', 'CHN', 'GER']
gold = [usgold, rusgold, gbrgold, chngold, gergold]
silver = [ussilver, russilver, gbrsilver, chnsilver, gersilver]
bronze = [usbronze, rusbronze, gbrbronze, chnbronze, gerbronze]

plt.plot(names, gold, linestyle = '-.', color='gold', label = 'Gold Medals', marker='o', linewidth=2.5)
plt.plot(names, silver, linestyle = ':', color='silver', label = 'Silver Medals', marker='o', linewidth=2.5, alpha=10)
plt.plot(names, bronze, linestyle = '--', color='brown', label = 'Bronze Medals', marker='o', linewidth=2.5)

plt.xlabel('Nationality')
plt.ylabel('Medals')

plt.legend()
plt.tight_layout()


plt.title('Medal Count')

plt.show()
```

##  -- Heat Map

##### The correlation heatmap ranges from -1 to +1 (scale on the right side). Values that are closer to 0 means there's no linear trend (no real relationship) between the two variables. A correlation closer to 1 means that it is more positively correlated; that is, as one increases so does the other (the stronger the relationship). Similarly, a correlation that is closer to -1 is one variable decreasing as the other increases (or vice versa), rather than both increasing or decreasing. Our conclusion to the heatmap confirms that height and weight are correlated (while other variables are not).

```python
#### Heat Map

plt.figure(figsize=(10,4)) 
sns.heatmap(athletes.corr().round(2), annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm', linewidths=5)    
plt.title('Correlation Matrix', fontsize = 15)

plt.show()
```

### -- Bar Chart

##### The bar chart shows the number of medals won by each nationality during the 2016 Olympics. Since there are over 200 different nationalities that participated, we decided to plot only the top 10 nationalities with the highest number of athletes competing and the number of medals they won. The x-axis shows the top 10 nationalities and the y-axis shows the number of gold, silver, and bronze medals won by each nationality, so for example, USA won the most gold medals during the 2016 Olympics while the UK won the most silver medals.


```python
#Bar Chart

counts = collections.Counter(athletes.nationality)
top10countries = [elem for elem, _ in counts.most_common(10)]

df_top10 = athletes[athletes['nationality'].isin(top10countries)]

df = df_top10.groupby('nationality')[['gold','silver','bronze']].sum()
index = np.arange(10)
bar_width = 0.2

r1 = plt.bar(index, df['gold'], bar_width, color ='gold', label='Gold')
r2 = plt.bar(index + bar_width, df['silver'], bar_width, color = 'silver', label = 'Silver')
r3 = plt.bar(index + bar_width + bar_width, df['bronze'], bar_width, color = 'brown',  label = 'Bronze')

plt.xticks(index + bar_width, df.index, color = 'black', rotation = 45, horizontalalignment = 'right')

plt.legend()
plt.tight_layout()
plt.show()
```

### -- Bubble Chart

##### The x-axis shows us that the further down you go, the larger the teams get per a given nationality. USA, Germany, and Brazil are the top three with the largest number of players. The further up you go on the y-axis, you get more medals won for each nationality. It seems like there is a direct correlation for the United States where the more players you have, the higher the chance the country has of getting more medals. Brazil is an outlier because even though they have the 2nd largest number of players, they did not win as many medals. The graph does have a positive correlation. 

```python
#Bubble Chart
nationalities = list(data['nationality'].unique()) #extract nationalities from data
athletes_per_nationality = [len(data[data['nationality'] == n]) for n in nationalities] #extract gold medals from data
gold_medals_per_nationality = [data[data['nationality'] == n]['gold'].sum() for n in nationalities] #extract gold medals from data
silver_medals_per_nationality = [data[data['nationality'] == n]['silver'].sum() for n in nationalities] #extract silver medals from data
bronze_medals_per_nationality = [data[data['nationality'] == n]['bronze'].sum() for n in nationalities] #extract bronze medals from data
total_medals_per_nationality = [sum(i) for i in zip(gold_medals_per_nationality, silver_medals_per_nationality, bronze_medals_per_nationality)] #combine medals to get total medals from data
medals_per_player_per_nationality = [total_medals_per_nationality[i] / athletes_per_nationality[i] for i in range(len(nationalities))] #calculate number of medals per player

fig, ax = plt.subplots() #initialize plot
bubble_sizes = [2000*x for x in medals_per_player_per_nationality] #create bubble sizes list
ax.scatter(athletes_per_nationality, total_medals_per_nationality, s=bubble_sizes, alpha=0.5) #make bubble plot as scatter plot
#iterate over all of the bubbles
for i, txt in enumerate(nationalities):
    #if the bubble is large enough
    if athletes_per_nationality[i] > 150 or total_medals_per_nationality[i] > 25 or medals_per_player_per_nationality[i] > 0.35:
        #add annotation label with nationality in bubble
        ax.annotate(txt, (athletes_per_nationality[i], total_medals_per_nationality[i]), ha='center', va='center')
        
plt.xlabel("Total Number of Athletes for each Nationality") #label x axis
plt.ylabel("Total Number of Medals for each Nationality") #label y axis
plt.title("Olympic Results for each Nationality") #add plot title

plt.tight_layout()
plt.show() #display bubble plot
```
